```python
regional_summary = (
    merged_df
    .groupby("Region")
    .agg(
        Avg_Recommended_Price=("Recommended_selling_price", "mean"),
        Target_Margin=("Target_margin", "mean")
    )
    .sort_values("Avg_Recommended_Price", ascending=False)
)

regional_summary
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

bars = plt.bar(
    plot_df["Region"],
    plot_df["Avg_Recommended_Price"]
)

plt.title("Average Recommended Price by Region", fontsize=14)
plt.xlabel("Region", fontsize=11)
plt.ylabel("Average Price ()", fontsize=11)

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f"{height:.2f}",
        ha="center",
        va="bottom",
        fontsize=10
    )

plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
from scipy import stats

uk_prices = merged_df.loc[merged_df["Region"] == "UK", "Recommended_selling_price"]
germany_prices = merged_df.loc[merged_df["Region"] == "Germany", "Recommended_selling_price"]
```

```python
t_stat, p_value = stats.ttest_ind(
    uk_prices,
    germany_prices,
    equal_var=False  # safer for real data
)

t_stat, p_value
```

```python
if p_value < 0.05:
    print("Statistically significant price difference")
else:
    print("No statistically significant difference")
```

```python
from scipy.optimize import minimize

def profit(price):
    cost = 2.5
    demand = 1000 - 120 * price  # simplified demand curve
    return -(price - cost) * demand  # negative for minimization
```

```python
merged_df["Is_Premium"] = (
    merged_df["Price_Tier"] == "Premium"
).astype(int)
```

```python
features = [
    "Brand_Strength_Score",
    "Target_margin",
    "Base_Cost"
]

X = merged_df[features]
y = merged_df["Is_Premium"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = log_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
premium_probabilities = log_model.predict_proba(X_test)
premium_probabilities[:5]
```

```python
premium_prob = premium_probabilities[:, 1]
premium_prob
```

```python
results_df = X_test.copy()
results_df["Actual_Is_Premium"] = y_test.values
results_df["Predicted_Premium_Prob"] = premium_prob

results_df.head()
```

```python
def premium_decision(prob):
    if prob >= 0.75:
        return "Strong Premium Candidate"
    elif prob >= 0.5:
        return "Borderline Premium"
    else:
        return "Standard / Value"

results_df["Premium_Decision"] = results_df["Predicted_Premium_Prob"].apply(premium_decision)
results_df.head()
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

fpr, tpr, _ = roc_curve(y_test, premium_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(7,5))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")
plt.plot([0,1], [0,1], linestyle="--")

plt.title("ROC Curve  Premium Classification", fontsize=14)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid(alpha=0.5)
plt.tight_layout()
plt.show()
```
